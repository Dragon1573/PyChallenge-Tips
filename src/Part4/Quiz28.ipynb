&emsp;&emsp;关卡图片`bell.png`是一张全彩色图片，我们需要将它的绿色（G）通道提取出来。

```python
from io import BytesIO
from PIL import Image
from IPython.display import display
```

```python
# 获取图片
response = get(
    'http://www.pythonchallenge.com/pc/ring/bell.png',
    headers={'Authorization': 'Basic cmVwZWF0OnN3aXRjaA=='}
)
# 打开图片
image = Image.open(BytesIO(response.content))
# 展示图片
display(image)
# 提取数据
data = list(image.split()[1].getdata())
# 将数据成对相减，忽略其中差为42的数据对
differences = list(filter(lambda x : x != 42, [abs(x - y) for (x, y) in zip(data[0::2], data[1::2])]))
# 以Unicode方式解码
print(bytes(differences).decode())
```
