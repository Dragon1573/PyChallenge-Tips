&emsp;&emsp;放大图片，我们发现这张图片是由2张原始图片以国际象棋盘的交错排列方式拼合而成的。我们使用`PIL`将图片进行分离。

```python
from requests import post
from PIL import Image
from io import BytesIO
```

```python
# 爬取原图
response = post(
    'http://www.pythonchallenge.com/pc/return/cave.jpg',
    headers={'Authorization': 'Basic aHVnZTpmaWxl'}
)
# 打开原图
raw = Image.open(BytesIO(response.content))
# 获取原图大小
size = [k // 2 for k in raw.size]
# 创建新图片
odd = Image.new('RGBA', size)
even = Image.new('RGBA', size)
# 像素分离
for x in range(raw.size[0]):
    for y in range(raw.size[1]):
        if (x + y) % 2 == 0:
            even.putpixel((x // 2, y // 2), raw.getpixel((x, y)))
        else:
            odd.putpixel((x // 2, y // 2), raw.getpixel((x, y)))
# 亮度提升
odd.point(lambda x: x + 100)
even.point(lambda x: x + 500)
# 展示图片
display(odd)
display(even)
```
