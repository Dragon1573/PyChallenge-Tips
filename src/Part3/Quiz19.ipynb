```python
from requests import get
from bs4 import BeautifulSoup as Soup
```

```python
response = get(
    'http://www.pythonchallenge.com/pc/hex/bin.html',
    headers={'Authorization': 'Basic YnV0dGVyOmZseQ=='}
)
response = Soup(response.text, features='html5lib')
response = str(response.contents[1])
response = response.splitlines()
print('\n'.join(response[1:15]))
print('...')
```

&emsp;&emsp;源代码中的注释包含了一封来自列奥波尔得·莫扎特的电子邮件，老莫向我们询问“打开**附件**”的意思，与我们说明他的设备出现故障，并将一段以`base64`格式编码的波形音频文件符在了邮件末尾。所以，我们需要帮老莫提取`base64`格式的附件内容，并还原为音频文件`indian.wav`。

```python
from urllib.parse import unquote_to_bytes
from base64 import decodebytes
from io import BytesIO
import wave
from IPython.display import Audio
```

```python
attachment = ''.join(response[14:len(response) - 3])
attachment = decodebytes(unquote_to_bytes(attachment))
display(Audio(attachment))
```

&emsp;&emsp;音频中出现了关键词`sorry`，前往<http://www.pythonchallenge.com/pc/hex/sorry.html>。

```python
response = get(
    'http://www.pythonchallenge.com/pc/hex/sorry.html',
    headers={'Authorization': 'Basic YnV0dGVyOmZseQ=='}
)
print(response.text.strip())
```

&emsp;&emsp;`(・∀・(・∀・(・∀・*)`网址看上去不对。`out of order`还有**杂乱无章**的含义，将音频反转试试？

```python
raw = wave.open(BytesIO(attachment), 'rb')
new = wave.open(BytesIO(), 'wb')
new.setnchannels(raw.getnchannels())
new.setsampwidth(raw.getsampwidth())
new.setframerate(raw.getframerate())
for k in range(raw.getnframes()):
    new.writeframes(raw.readframes(k)[::-1])
display(Audio(new._file.getvalue()))
```

&emsp;&emsp;新的音频中出现了关键词`idiot`，再试试<http://www.pythonchallenge.com/pc/hex/idiot.html>。
