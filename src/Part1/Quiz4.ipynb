&emsp;&emsp;由于提示的格式是固定的，我们尝试用正则表达式提取下级链接、实现自动跳转。

```python
from requests import get
from re import search
```

```python
index = "12345"
PREFIX = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing="
hops = 0
while True:
    response = get(PREFIX + index)
    print("Hops %d: %s" % (hops, response.text))
    hops += 1
    try:
        # 此处使用 try catch 语句块是由于跳转链中存在1种特殊情况
        # 即提示中不存在数字，此时需要进行异常处理
        index = search(r'nothing is ([0-9]+)', response.text).group(1)
    except AttributeError:
        if search('Yes', response.text):
            index = str(int(index) // 2)
        else:
            break
```

&emsp;&emsp;通过250次重定向跳转后，我们终于找到了下一关的链接：<http://www.pythonchallenge.com/pc/def/peak.html>。
